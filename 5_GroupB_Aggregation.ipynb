```python
import pandas as pd
```

```python
data = {
    'Department': ['HR', 'IT', 'HR', 'Finance', 'IT', 'Finance'],
    'Employee': ['John', 'Anna', 'Peter', 'Linda', 'Bob', 'Sam'],
    'Salary': [60000, 80000, 65000, 70000, 90000, 75000],
    'Bonus': [5000, 10000, 7000, 8000, 12000, 9000]
}
df = pd.DataFrame(data)
```

## Aggregation

```python
df["Salary"].agg(['mean', 'std', 'count', 'median'])
```

### What do these statistics mean and how are they calculated?
- **mean**: The average value. Shows the typical salary in the data.
  - **Formula:** $\text{mean} = \frac{\text{sum of all values}}{\text{number of values}}$
- **std** (standard deviation): Measures how much the salaries vary from the mean. High std means more spread out values.
  - **Formula:** $\text{std} = \sqrt{\frac{1}{N} \sum_{i=1}^N (x_i - \text{mean})^2}$
- **median**: The middle value when salaries are sorted. Useful if there are outliers, as it is not affected by extreme values.
  - **How:** Sort all values; the median is the center value (or average of two center values if count is even).

Use these to quickly understand the center, spread, and distribution of your salary data.

```python
# Import numpy and pandas libraries
import numpy as np
import pandas as pd
```

```python
# Create a DataFrame for employees with id, name, and department
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})
```

```python
# Display the employees DataFrame
employees
```

```python
# Create a DataFrame for salaries with employee_id, salary, and bonus
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
# Display the salaries DataFrame
salaries
```

#### Merge two dataframe

```python
pd.merge(employees, salaries, on="employee_id", how="inner")
```

```python
pd.merge(employees,salaries,on='employee_id',how ='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='right')
```

#### Concatination of 2 dataframes

```python
# Create two DataFrames for demonstrating concatenation

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
# Display the first DataFrame for concat example
df1
```

```python
# Display the second DataFrame for concat example
df2
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2], axis=1)
```

#### Joining 2 data frames

```python
# Create two DataFrames with custom indices for join demonstration

df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
# Display the first DataFrame for join example
df1
```

```python
# Display the second DataFrame for join example
df2
```

```python
# Outer join on index: combines all indices from both DataFrames
# Missing values are filled with NaN

df1.join(df2, how="outer")
```

```python
# Join df2 with df1 on index (default is left join)
df2.join(df1)
```

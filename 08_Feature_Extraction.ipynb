```python
df["Total_Time"] = df["Title"].apply(lambda text : re.search(r'[A-Za-z]{3}\s\d{4}\s-\s[A-Za-z]{3}\s\d{4}', text).group())
```

```python
df.head()
```

```python
def calculate_months(text):
    # Split dates
    start_str, end_str = text.split(" - ")
    
    # Convert to datetime
    start = datetime.strptime(start_str, "%b %Y")
    end = datetime.strptime(end_str, "%b %Y")
    
    # Calculate total months
    months = (end.year - start.year) * 12 + (end.month - start.month)
    return months
```

```python
df["Months"] = df["Total_Time"].apply(calculate_months)
```

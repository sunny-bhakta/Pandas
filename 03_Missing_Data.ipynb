```python
import numpy as np
import pandas as pd
```

## Find missing data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().any()
```

```python
df
```

```python
df.dropna() # As first rows has non null value so rest drops
```

```python
df
```

```python
# ✔ Rows with all NaN values will be removed
# ✔ Rows with even one valid value will stay
df.dropna(thresh=1) 
```

## Filling missing data

```python
df
```

```python
df.fillna(0)
```

```python
df
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
